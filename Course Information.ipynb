# Applied Data Science for Data Analysts

## Welcome

Welcome to *Applied Data Science for Data Analysts* by Databricks. We're thrilled you can join us.

## Course Materials

*Coursera*

The majority of the intructional materials for this course can be found in Coursera. This includes videos, reading activities, discussion activities, knowledge checks, and quizzes. 

<img alt="Side Note" title="Side Note" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.05em; transform:rotate(15deg)" src="https://files.training.databricks.com/static/images/icon-note.webp"/> **Please use Coursera to progress through this course**.

*Databricks*

The material in this location needs to be run and used in the Databricks platform. This includes demos, labs, and projects. 

<img alt="Side Note" title="Side Note" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.05em; transform:rotate(15deg)" src="https://files.training.databricks.com/static/images/icon-note.webp"/> **Please follow along and complete hands-on activities in Databricks**.

## Structure

The Databricks course materials are organized as follows.

We have five modules with Databricks labs and projects:

1. **Module 1**
2. **Module 2**
3. **Module 3**
4. **Module 4**
4. **Module 5**

Each of these modules has their own folder within our course. Within each module's folder, there are subfolders for each lesson with Databricks content.

For example, **Module 3** has the following folders:

1. **Lesson 3.1**
2. **Lesson 3.2**

Each of these lessons contain Databricks notebooks for lecture demos and for completing labs.

<img alt="Side Note" title="Side Note" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.05em; transform:rotate(15deg)" src="https://files.training.databricks.com/static/images/icon-note.webp"/> The **Includes** folder facilitates the set up of the course (data, etc.). You do not need to interact with the **Includes** folder.

## Getting Started with Databricks

If you need an overview of Databricks content, please check out **Module 1**. The Coursera videos and associated Databricks content contain overviews for everything you'll need to know.

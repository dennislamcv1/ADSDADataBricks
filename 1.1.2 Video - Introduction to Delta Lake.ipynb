## Video Content

Use the below cells to follow long with the video.

The below command cell will display the first 1000 rows of data in the **`ht_users`** table in the **`adsda`** database.

```
%sql
SELECT * FROM adsda.ht_users
```

The below command cell will create a table containing only the athletes from **`dsfd.ht_users`**. It will name the new table **`ht_users_athletes`** and put it in the **`dsfda`** database. It will store the data for the table at `/dsfda/ht-users-athletes`.

```
%sql
CREATE OR REPLACE TABLE adsda.ht_users_athletes
USING DELTA LOCATION '/adsda/ht-users-athletes'
AS (
  SELECT *
  FROM adsda.ht_users
  WHERE lifestyle = 'Athlete'
)
```

We can also run SQL queries on tables and return the results as a Spark DataFrame.

```
ht_users_non_athletes_df = spark.sql(
  "SELECT * FROM adsda.ht_users WHERE lifestyle != 'Athlete'"
)
```

In order to view the data in DataFrames, we need to run the `display()` function.

```
display(ht_users_non_athletes_df)
```

Sometimes we might want to create tables from DataFrames. First, we need to write out the DataFrame in the Delta format.

```
ht_users_non_athletes_df.write.format("delta").mode("overwrite").save("/adsda/ht-users-non-athletes")
```

Next, we registed the table with location associated with DataFrame we just saved.

```
%sql
CREATE TABLE adsda.ht_users_non_athletes
USING DELTA LOCATION '/adsda/ht-users-non-athletes'
```

The below command cell confirms that we saved our DataFrame as a Delta table.

```
%sql
SELECT * FROM adsda.ht_users_non_athletes
```
